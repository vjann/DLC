```python
#turns seconds into a string format ffmpeg uses
def seconds_formatter(sec):
    x = "0" + str(datetime.timedelta(seconds=sec))[:-3]
    return x
```

```python
videos = [r"D:\DCIM\100GOPRO\2019.07.16_CC42970_N3_V3_2.7k_120FPS.MP4", r"D:\DCIM\100GOPRO\2019.07.16_CC42970_N2_V3_2.7k_120FPS.MP4", r"D:\DCIM\100GOPRO\2019.07.16_CC42970_N3_V2_2.7k_120FPS.MP4"]
ranges = [range(11, 485, 10), range(7, 445, 10), range(44, 300, 10)]
for video_path, r in zip(videos, ranges):
    video_directory = video_path.rsplit('\\', 1)[0]
    video_name = video_path[:-4].rsplit('\\', 1)[1]
    bat_file= video_directory + r"\{}_create_videos.bat".format(video_name)
    target_directory = video_directory + r"\{}_cut".format(video_name)
    print(bat_file)
    #Open up new batch file
    script_writer= open(bat_file,"w")

    script_writer.write('if not exist "{}" mkdir {} && '.format(target_directory, target_directory))

    #write series of shell commands to create videos for each attempt, connected by '&&' for the batch file
    vids = []
    for time in r:
        start_time = seconds_formatter(max(0, time - 0.5))
        new_vid_path = target_directory + "\\" + str(np.round(time, 3)) + ".mp4"
        print(new_vid_path)
        cmd = "ffmpeg -i {0}.mp4 -ss {1} -t {2} -c copy {3}".format(video_name, start_time, 7, new_vid_path)
        vids.append(new_vid_path)
        script_writer.write(cmd + " && ")
    script_writer.close()

    #open and write a list of videos to concatenate (.txt file) for ffmpeg to use
    list_file = video_directory + r"\{}_video_list.txt".format(video_name)
    list_writer = open(list_file, "w+")
    for v in vids:
        list_writer.write('file ' + "'" + v + "'" + "\n")
    list_writer.close()

    #open batch file again and write command to produce a concatenated video, the final product
    script_writer = open(bat_file, "a+")
    script_writer.write("ffmpeg -f concat -safe 0 -i {0} -c copy {1}.mp4".format(list_file, video_name + "_spliced"))
    script_writer.close()
```
